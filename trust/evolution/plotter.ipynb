```python
import matplotlib.pyplot as plt
import numpy as np

# Heights for the horizontal lines
line_heights = [3701, 4521, 5384]
#line_heights = [587, 644, 695]
line_labels = ['Always', 'Random', 'Never']
line_colors = ['red', 'green', 'blue']

# Data points
x_values = np.array([1, 2, 3, 4, 5])
y_values = np.array([4339, 4058, 3758, 4376, 4930])
# y_values = np.array([681, 606, 624, 643, 659])

# Create the figure and axis
plt.figure(figsize=(8, 5))

# Plot horizontal lines
for height, label, color in zip(line_heights, line_labels, line_colors):
    plt.axhline(y=height, color=color, linestyle='--', linewidth=1, label=label)

# Plot data points
plt.plot(x_values, y_values, marker='o', color='blue', label='Data Points')

# Labels and title
plt.title('Alice evaluation')
plt.xlabel('Game number')
plt.ylabel('Ticks per game')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
